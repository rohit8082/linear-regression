```python
import numpy as np    #for numerical operations
import pandas as pd   # for dataframes
```

```python
def read__csv():
 df1=pd.DataFrame(pd.read_csv("/content/drive/MyDrive/data_YesBank_StockPrices.csv"))
 return df1
```

```python
read__csv()
```

```python
from sklearn.model_selection import train_test_split            #splitting training data and test data from dataframe
X1=pd.DataFrame(read__csv().drop(['Close','Date'],axis=1))
X2=pd.DataFrame(read__csv()['Close'])
```

```python
X1_train,X1_test,X2_train,X2_test=train_test_split(X1,X2,test_size=0.25)      # actuall splitting data , training =75%  ,  test=25%
```

```python
from sklearn.linear_model import LinearRegression        #importing linear regression model
```

```python
LR=LinearRegression()         # object of LR
```

```python
LR.fit(X1,X2)          # Fitting LR model
```

```python
r_sq=LR.score(X1,X2)   # finding variance
r_sq
```

```python
 y_pred = LR.predict(X1)      # predicting the values of (close)
```

```python
y_pred
```

```python
# plot for residual error
import matplotlib.pyplot as plt
# setting plot style
plt.style.use('fivethirtyeight')

# plotting residual errors in training data
plt.scatter(LR.predict(X1_train),
            LR.predict(X1_train) - X2_train,
            color="green", s=10,
            label='Train data')

# plotting residual errors in test data
plt.scatter(LR.predict(X1_test),
            LR.predict(X1_test) - X2_test,
            color="blue", s=10,
            label='Test data')

# plotting line for zero residual error
plt.hlines(y=0, xmin=0, xmax=50, linewidth=2)

# plotting legend
plt.legend(loc='upper right')

# plot title
plt.title("Residual errors")

# method call for showing the plot
plt.show()
```
